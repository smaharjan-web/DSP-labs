```python
import numpy as np
def func(n):
    return (int)((1/np.sqrt(5)) * ((((1 + np.sqrt(5))/2) ** n) - (((1 - np.sqrt(5))/2) ** n)))
```

```python
import time
import matplotlib.pyplot as plt

my_time = []
my_num = []

for n in range(0, 30):
    
    start = time.time() #start timer

    out = func(n) #obtain elements using above described function
    
    end = time.time() #end timer
    
    t = end-start #calculate total time taken
    
    my_time.append(t) #append times
    my_num.append(n) #append numbers in range

#array conversion
j = np.array(my_num)
k = np.array(my_time)

#plot graph 1
plt.plot(j,k)
plt.xlabel('n')
plt.ylabel('time') 
plt.title('Graph without recursion, using f(n)')
plt.show()
```

```python
my_newtime = []

def recursion(n):
    
    if n==0:
        return 0

    elif n==1:    
        return 1    
        
    else:
        return (recursion(n-1) + recursion(n-2))
            
```

```python
for n in range(0,30):
        
    startx = time.time()      #start timer
    elements = recursion(n)      #obtain elements using recursion function defined above
    finishx = time.time()     #end timer
        
    timer = finishx - startx #calculate the total time taken
        
    my_newtime.append(timer) #append times
    my_num.append(n) #append numbers in range

print(my_newtime)
l = np.array(my_newtime) #array conversion
```

```python
#plot graph 2
plt.plot(j,l)
plt.xlabel('n')
plt.ylabel('time(sec)')
plt.title('Graph using recursion')
plt.show()
```

```python
# Graph 1 and 2 combined
plt.plot(j,k)
plt.plot(j,l)
plt.xlabel('n')
plt.ylabel('time(sec)')
plt.title('Graphs using f(n) directly(blue) and using recursion(orange)')
```

```python
#impulse response of the given circuit 
def system(t):
    return (np.power(10,6)*np.exp(-np.power(10,6)*t))
```

```python
#input vector is 1000 samples long
input_vector = []
for k in range(0,1000):
    input_vector.append(1)

print (input_vector[0:5])
```

```python
#impulse reponse for each t = nT WHERE T = 10^(-8)
impulse_vector = []

for x in range(0,1000):
    impulse_vector.append(system(x * (10 ** -8))) #np.power gave errors so used ** instead

print(impulse_vector[0:5])
```

```python
#response of the circuit

output = np.convolve(input_vector,impulse_vector)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def func(t):
    return np.exp(-t)

def forward(n,T):
    return T*((1-(1*T))**(n-1))

#def backward(n,T):
    #return T*(1+T)*((1+T)**n)

def backward(n,T):
    return ((1+T)**(-n))
    

x = 2.1
y = 1.5
z = 0.8
horizontal = np.arange(0.1,11,0.001)
plt.plot(horizontal,func(horizontal))

```

```python
n = np.arange(1,16,1)
plt.plot(n,forward(n,x))
```

```python
plt.plot(horizontal,func(horizontal))
plt.plot(n,backward(n,x))
```

```python
plt.plot(horizontal,func(horizontal))
plt.plot(n,forward(n,y))
```

```python
plt.plot(horizontal,func(horizontal))
plt.plot(n,backward(n,y))
```

```python
plt.plot(horizontal,func(horizontal))
plt.plot(n,forward(n,z))
```

```python
plt.plot(horizontal,func(horizontal))
plt.plot(n,backward(n,z))
```

```python
# Four axes, returned as a 2-d array
f, axarr = plt.subplots(3, 2)

axarr[0, 0].plot(horizontal,func(horizontal))
axarr[0, 0].plot(n,forward(n,x))

axarr[0, 0].set_title('Forwards difference, T=2.1, Pole = -1.1')

axarr[0, 1].plot(horizontal,func(horizontal))
axarr[0, 1].plot(n,backward(n,x))

axarr[0, 1].set_title('Backwards difference, T=2.1, Pole = 0.3')


axarr[1, 0].plot(horizontal,func(horizontal))
axarr[1, 0].plot(n,forward(n,y))

axarr[1, 0].set_title('Forwards difference, T=1.5, Pole = -0.5')

axarr[1, 1].plot(horizontal,func(horizontal))
axarr[1, 1].plot(n,backward(n,y))

axarr[1, 1].set_title('Backwards difference, T=1.5, Pole = 0.4')


axarr[2, 0].plot(horizontal,func(horizontal))
axarr[2, 0].plot(n,forward(n,z))

axarr[2, 0].set_title('Forwards difference, T=0.8, Pole = 0.2')

axarr[2, 1].plot(horizontal,func(horizontal))
axarr[2, 1].plot(n,backward(n,z))

axarr[2, 1].set_title('Backwards difference, T=0.8, Pole = 0.6')

f.subplots_adjust(hspace=0.8)
f.subplots_adjust(wspace=0.5)
f.set_figheight(15)
f.set_figwidth(15)
```
